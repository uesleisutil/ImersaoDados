### 3.0. Análise exploratória dos dados.

Com a seção abaixo é possível concluir que existem:

* 2 tipos de tratamento;
* 3 tempos;
* 2 doses;
* 3289 drogas.

```python
# Filtragem dos dados: Utilizar as colunas presentes no *dataframe*, com excessão das colunas dos genes.
filtered_df = [str(column) for column in df.columns if not (str(column).startswith('g-') or str(column).startswith('c-') or column == 'id')]

# Cria uma nova variável apenas com os valores únicos destas colunos.abs
unique = {column: [len(df[column].value_counts())] for column in filtered_df}

# Transforma a variável em dataframe.
unique = pd.DataFrame(unique)
unique
```

#### 3.1. Investigar a distribuição dos dados através de gráficos de barras.

A seção abaixo demonstra que:
* Existe um desbalanceamento entre os dados *com_controle* e *com_droga*;
* Os dados com *48* horas se sobressaem ante os daos de *24* e *48* horas.
* A dose *D1* se repete mais que a dose *D2*. 


```python
# Amostra dos dados.
samples = [df[column].value_counts() for column in filtered_df]

# Recursos para plotar as figuras.
fig, axs = subplots(1, len(samples), figsize=(25,5))

# Plota, em barras, os primeiros 30 dados das tabelas.
for i, samples in enumerate(samples):
    title = filtered_df[i].capitalize()
    samples.nlargest(30).sort_values().plot(kind='bar', ax=axs[i], title=title)
```

### 3.2. Investigar o desbalanceameno da coluna *tratamento*.

- *com_droga* representa ~92% dos dados;
- *com_controle* representa ~8% dos dados.

Ao se comparar os testes por droga, é possível verificar que a presença do controle é maior, com *1750* testes. Portanto, o baixo percentual de utilização do controle se justifica devido ao alto número de drogas diferentes sendo testadas.

```python
# Cria uma nova variável.
proportion = pd.DataFrame(df["tratamento"].value_counts())

# Plota em barra.
proportion.plot.bar(title="Tratamento e Controle", subplots=True)
```

### 2.0. Utilizar o seaborn.

```python
import seaborn as sns
import matplotlib. pyplot as plt
sns.set()
```

```python
compounds = df['composto'].value_counts().index[0:5]

plt.figure(figsize=(10,8))
sns.countplot(x='composto', data=df.query('composto in @compounds'),
              order=compounds.value_counts(ascending=True).index)
ax = plt.gca()
ax.set_title('Top 5 Compostos', fontsize=15)
ax.tick_params(labelsize=14, length=2, color='0.2', labelcolor='0.2',direction='in') 
plt.xlabel('Compostos',labelpad=18,size=15)
plt.ylabel('Frequência',labelpad=33,size=15)
plt.show()
        
```
